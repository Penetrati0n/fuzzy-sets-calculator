# Core

```python
import numpy as np
import pandas as pd

MATRIXS = {}

def parse_matrix(name: str, data: str):
    global MATRIXS
    MATRIXS[name] = np.matrix(data.replace(',', '.').strip().replace('\n', ';'))

def get_matrix(name: str):
    global MATRIXS
    return MATRIXS[name]

# A в B.
def AinB(a, b):
    return (a <= b).sum() == np.multiply(*a.shape)

# B в A.
def BinA(a, b):
    return (a >= b).sum() == np.multiply(*a.shape)

# A == B.
def AeqB(a, b):
    return (a == b).sum() == np.multiply(*a.shape)

# A != B.
def AneqB(a, b):
    return (a != b).sum() == np.multiply(*a.shape)

# Вложенность.
def task1(name1: str, name2: str):
    matrix_a = get_matrix(name1)
    matrix_b = get_matrix(name2)
    res = [0, 0, 0, 0]
    key_element = None
    if AinB(matrix_a, matrix_b): res[0] = 1
    elif BinA(matrix_a, matrix_b): res[1] = 1
    elif AeqB(matrix_a, matrix_b): res[2] = 1
    else: res[3] = 1
    if res[3]:
        new_matrix = matrix_a >= matrix_b
        if new_matrix.sum() * 2 > np.multiply(*new_matrix.shape):
            new_matrix = np.logical_not(new_matrix)
        key_elements = np.where(new_matrix)
        key_element = (key_elements[0][0], key_elements[1][0])
    return tuple(res + [key_element])

# Разложение.
def task2(name: str):
    matrix = get_matrix(name)
    uniq = np.unique(np.array(matrix))
    res = {}
    for i in np.sort(uniq[np.where(uniq > 0)]):
        res[i] = pd.DataFrame((matrix >= i).astype(int))
    return res
```

```python
# @title Разложение матрицы A на уровни
# ------------------------------------------------------------------------------
# Тут ничего менять не нужно.
# ------------------------------------------------------------------------------

ans = task2('A')
for l in ans:
    df = ans[l]
    print('Уровень\t{}'.format(l).replace('.', ','), end='\t')
    print(df.to_string(header=False, index=False).replace('.', ',').replace(' ', '\t').replace('\n', '\n\t\t'))
    for _ in range(10 - df.shape[0] + 1):
        print('\t' * df.shape[0])
```
